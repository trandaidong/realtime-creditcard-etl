```python
import os
import time
import random
from confluent_kafka import Producer

# Khởi tạo producer
conf = {
    'bootstrap.servers': 'localhost:9092'
}
producer = Producer(**conf)

topic = 'transactions' # Cần khỏi động kafka broker và tạo ra topic trước
 
def delivery_report(err, msg):
    if err is not None:
        print('Delivery failed:', err)
    else:
        print(f'Message delivered to {msg.topic()} [{msg.partition()}]')


path = os.path.abspath(os.path.join(os.getcwd(), '..', 'data', 'data.csv'))

with open(path, 'r') as f:
    header = next(f)  # bỏ dòng đầu
    for line in f:
        producer.produce(topic, value=line.strip(), callback=delivery_report)
        producer.poll(0.1)
        time.sleep(random.uniform(1, 3))  # ngẫu nhiên từ 1–3 giây

# Đảm bảo gửi hết
producer.flush()
```
